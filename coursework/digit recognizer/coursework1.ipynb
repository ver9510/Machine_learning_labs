Задача: распознавание черно-белой написанной от руки цифры. https://www.kaggle.com/c/digit-recognizer/data 
Входные данные: для каждой цифры задана матрица 28х28, каждое значение отражает яркость пиксела.  

```python
import pandas
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
```

```python
#загрузка датасета
def load_dataset(filename):
	csv_dataset = pandas.read_csv(filename, header=0).values
	dataset=csv_dataset
	return dataset
```

```python
dataset = load_dataset("../input/train.csv")
```

```python
class_train = dataset[:,0]
data_train = dataset[:,1:]
```

```python
test_dataset = load_dataset("../input/test.csv")
```

```python
forest = DecisionTreeClassifier(random_state=100)
forest = forest.fit( data_train, class_train )
```

```python
result = forest.predict(test_dataset)
```

```python
print("result with i")
i=1
for res in result:
    print(i,",",res)
    i=i+1
```
